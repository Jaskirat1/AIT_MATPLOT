```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
apple = np.array([8,25,7,9,2,55])
mango = np.array([4,9,3,3,566,44])
orange= np.array([2,6,8,50,66,44])
```

```python
plt.plot(apple)
plt.show()
```

types of markers in matplot 

```python
plt.plot(mango, marker='o')
plt.show()
```

```python
plt.plot(orange, marker = '*', ls = '--', c = 'r')
plt.show()
```

```python
plt.plot(apple, marker='o', mec='r', mfc='green', c = 'b', ls='--', lw =1)
plt.show()
```

```python
plt.plot(orange, marker='2', ms=20)
plt.xlabel("X-axis", size = 20)
plt.ylabel('Y-axis')
plt.title('graph ploting',loc='right')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, c=colors, cmap='viridis', s = sizes)
plt.colorbar()
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, c=colors, cmap='viridis', s = sizes, alpha=0.5)
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()
plt.figure(figsize=(4,4))
plt.plot(x,y)
plt.show()
```
