```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
a = np.array([33,23,245,6,68,90])
b = ['Apple', 'banana', 'cherry', 'orange', 'pineapple', 'dates']
c = [0.6,0,0,0,0,0.6]
plt.pie(a, labels=b, explode=c)
plt.show()
```

```python
plt.pie(a, labels=b, explode = c, autopct='%1.4f%%')
plt.show()
```

Bar Graphs

```python
x = np.random.rand(6)
y = np.random.randint(6)

x2 = np.array([23,34,56,23,57,90])
y2 = np.array([45,68,89,21,46,36])
plt.bar(x,y, width = 1.0)
plt.bar(x2,y2, width= 2.0)
plt.show()
```

For horizontal bar graphs we use height instead of using width

```python
plt.barh(x,y, color = 'yellow')
plt.barh(x2,y2, color = 'blue', height= 0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('BarGraph')
plt.show()
```

Histplot

```python
arr = np.array([45,89,81,80,79,65,42,76,84])
plt.hist(arr, color = 'red')
plt.title('Hist Plot')
plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.show()
```
